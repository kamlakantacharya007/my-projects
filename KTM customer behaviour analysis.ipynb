```python
#inporting libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read files
df = pd.read_csv("ktm.csv")
df
```

```python
df.info()
```

```python
df.columns
```

```python
#creating dummies and converting data in binary for logistic regression.
convert = ['Gender', 'Occupation', 'Phone Type', 'Current Bike',
       'Relationship', 'Response']
df_1 = pd.get_dummies(df,columns=convert,drop_first = True)
df_1
```

```python
df_1.columns
```

```python
#defining dependent & independent variable as x and y.
X=df_1.drop(['Response_Purchased'],axis=1).values
Y=df_1['Response_Purchased'].values
```

```python
#dropping ID coloumn 
df_2 =df_1.drop(['ID'],axis=1)
df_2
```

```python
#import logistic regression and dividing data inti train  & test
from sklearn.linear_model import LogisticRegression
glm = LogisticRegression()
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#fitting the model
glm.fit(X_train,Y_train)
```

```python
#model accuracy
from sklearn.metrics import accuracy_score
print('Logistic regression accuracy: {:.2f}'.format(accuracy_score(Y_test, glm.predict(X_test))))
```

```python
#confusion matrix
from sklearn import metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,glm.predict(X_test))
cm
```

```python
#plotting confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm = metrics.confusion_matrix(glm.predict(X_test),Y_test,[1,0])
sns.heatmap(cm, annot=True, fmt='.2f',xticklabels = ["Not purchased", "Purchased"] , yticklabels = ["Not purchased", "Purchased"] )
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Logistic Regression')
plt.savefig('logistic_regression')
```
