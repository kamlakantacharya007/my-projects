```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
#setting target users
target_users = ["@BJP4India", "@INCIndia", "@ShivSena"]
sentiments = []

#for target users creating compounds
for target in target_users:
    print(target)
    counter = 0
    
    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []
    
#exracting tweets from target users
    public_tweets = api.user_timeline(target,count=200)
    
    for tweet in public_tweets:
        
        results = analyzer.polarity_scores(tweet["text"])
        compound = results["compound"]
        pos = results["pos"]
        neu = results["neu"]
        neg = results["neg"]
        tweets_ago = counter
        
        sentiments.append({"Tweet": tweet["text"],"Political Parties": target,"Tweets Ago": counter,
                           "Date": tweet["created_at"], "Compound":compound,"Positive": pos,"Negative":neg,"Neutral": neu})
        
        counter = counter+1
```

```python
#storing data into features dataframe
sentiments_pd = pd.DataFrame.from_dict(sentiments)
sentiments_pd = sentiments_pd[['Political Parties', 'Tweets Ago', 'Date', 'Tweet', 
                               'Compound', 'Positive', 'Negative', 'Neutral']]
```

```python
#converting data into csv 
sentiments_pd.to_csv("Politics.csv",encoding="utf-8",index = False)
```

```python
#plottig tweets into different colours in scatter plot
orgs_colors_dict = {'@BJP4India':'lightblue','@INCIndia': 'green','@ShivSena': 'red'}

plt.scatter(sentiments_pd.groupby(["Political Parties"]).get_group("@BJP4India")["Tweets Ago"],
                sentiments_pd.groupby(["Political Parties"]).get_group("@BJP4India")["Compound"],
                  facecolors=orgs_colors_dict['@BJP4India'], edgecolors='black', label="BJP")
plt.scatter(sentiments_pd.groupby(["Political Parties"]).get_group("@INCIndia")["Tweets Ago"],
                sentiments_pd.groupby(["Political Parties"]).get_group("@INCIndia")["Compound"],
                  facecolors=orgs_colors_dict['@INCIndia'], edgecolors='black', label="INC")
plt.scatter(sentiments_pd.groupby(["Political Parties"]).get_group("@ShivSena")["Tweets Ago"],
                sentiments_pd.groupby(["Political Parties"]).get_group("@ShivSena")["Compound"],
                  facecolors=orgs_colors_dict['@ShivSena'], edgecolors='black', label="SHIVSENA")

now = datetime.now()
now = now.strftime("%m/%d/%y")
plt.title(f'Sentiment Analysis of Political Tweets ({now})')
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")

plt.xlim(100, 0)
plt.ylim(-1.0, 1.0)
yticks = [-1.0, -0.5, 0.0, 0.5, 1.0]
plt.yticks(yticks)

plt.legend(title="Political Sources", bbox_to_anchor=(1, 1), frameon=False)

plt.savefig("sentiment_analysis_of_political_tweets.png")
plt.show()
```

```python
#plotting twitter popularity in bar graph
x_axis = np.arange(sentiments_pd["Political Parties"].nunique())
tick_locations = [value+0.4 for value in x_axis]

plt.title(f'Overall Political Sentiment based on Twitter ({now})')
plt.xlabel("Political Source")
plt.ylabel("Tweet Polarity")

plt.bar(x_axis, sentiments_pd.groupby("Political Parties").mean()["Compound"],
       color=orgs_colors_dict.values(), align="edge", width =1)
plt.xticks(tick_locations, sentiments_pd["Political Parties"].unique())

plt.savefig("Bar_Sentiment.png")
plt.show()
```
